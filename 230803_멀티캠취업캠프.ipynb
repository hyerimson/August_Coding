2. OS별 폰트 설정

```python
import platform #platform : 운영체제가 돌아가는 시스템

from matplotlib import font_manager, rc  #rc : 어떤 라이브러리나 프로그램들이 시작할 때 자동으로 읽어들이는 초기값들
plt.rcParams['axes.unicode_minus'] = False  # '-'로 표시하게 해달라 #글꼴 없음으로 나타나는 경우가 가끔 있어서

if platform.system() == 'Darwin':   #iOS를 의미
    rc('font', family='AppleGothic')
elif platform.system() == 'Windows': #윈도우를 의미
    path = "c:/Windows/Fonts/malgun.ttf"
    font_name = font_manager.FontProperties(fname=path).get_name()
    rc('font', family=font_name)
else:
    font_path = './drive/MyDrive/NanumGothic.ttf'
    fontprop = font_manager.FontProperties(fname=font_path, size=10)
    font_name = fontprop.get_name()
    rc('font', family=font_name)
```

- 첫번째 방법은 필요할 때마다 따라가서 font 내용 추가
- 두번째 방법은 초기에 자동으로 설정해놓았으니 따라다니며 font_name 쓸 필요 없다
- 1번은 코랩, 2번은 로컬에서 주로 씀
