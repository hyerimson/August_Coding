##Streamlit
1. streamlit이 설치된 가상환경으로 진입힌다.
2. app.py 파일이 있는 폴더로 이동한다.
3. streamlit run app.py 명령어로 웹서버를 작동시킨다.

### streamlit 웹서버 종료 방법
    - streamlit을 실행시킨터미널(콘솔, powershell prompt)에서
    
    crtl + c 버튼을 누른다.

- 8502는 포트번호

### streamlit 웹서버는 app.py 파일에 있는 내용을 실행시켜 웹브라우저에 표출해준다.

### app.py 파일에는 streamlit에서 제공하는 명령어들을 사용하여 원하는 내용을 웹브라우저에 표출시킬 수 있다.

### streamlit에서 웹브라우저에 뭔가를 표시하기 위해서는 streamlit 명령어를 사용해야 한다.

* streamlit 명령어는 파이썬기본 문법을사용할 수 있다.
* 파이썬 코드들을 수용할 수 있다.

---
<최소한 알고있으면 좋은 streamlit 명령어>
- st.dataframe
 - 데이터 프레임을 대화형 테이블로 표시
      
        st.dataframe(data=None, width=None, height=None, *,
        use_container_width=False,
        hide_index=None,
        column_order=None,
        column_config=None)
 - 예시
        import streamlit as st
        import pandas as pd
        import numpy as np

        df = pd.DataFrame(
        np.random.randn(50, 20),
        columns=('col %d' % i for i in range(20)))

        st.dataframe(df)  # Same as st.write(df)

- st.pyplot
 - matplotlib.pyplot 그림을 표시
        st.pyplot(fig=None,
        clear_figure=None,
        use_container_width=True, **kwargs)
 - 예시
        arr = np.random.normal(1, 1, size=100)
        fig, ax = plt.subplots()
        ax.hist(arr, bins=20)

        st.pyplot(fig)

- st.button
 - 버튼 위젯을 표시
        st.button(label, key=None,
        help=None, on_click=None, args=None,
        kwargs=None, *, type="secondary",
        disabled=False,
        use_container_width=False)
 - 예시
        import streamlit as st

        if st.button('Say hello'):
            st.write('Why hello there')
        else:
            st.write('Goodbye')

- st.text_input
 - 텍스트 데이터를 입력하고 싶은 경우
        st.text_input(label, value="",
        max_chars=None, key=None,
        type="default", help=None,
        autocomplete=None, on_change=None,
        args=None, kwargs=None, *,
        placeholder=None, disabled=False,
        label_visibility="visible")
 - 예시
        title = st.text_input('Movie title', 'Life of Brian')
        
        st.write('The current movie title is', title)

- st.header/title/write
 - title : 제목 형식으로 텍스트를 표시
        st.title('This is a title')
        st.title('A title with _italics_ :blue[colors] and emojis :sunglasses:')
 - header : 헤더 형식으로 텍스트를 표시
        st.header('This is a header')
        st.header('A header with _italics_ :blue[colors] and emojis :sunglasses:')
 - write : 인수 쓰기
        st.write('Hello, *World!* :sunglasses:')

app.py : 웹서버를 리스닝(계속 지켜본다)하며, 페이지 변경 요청이 있을 때, 페이지 구분을 해준다.

page 폴더의 파일들 : app.py로부터 호출을 받아서 클라이언트 역할을 하며, utils 폴더에 작성되는 모듈로부터 함수를 호출한다.

utils 폴더의 파일들 : page 폴더에 있는 파일들의 모듈로써, 파이썬 코드로만 작성된 경우가 대부분이며 실제 코드를 작성해야 하는 부분이다.

app - page 폴더 파일 (클라이언트 - 모듈)

page 폴더 파일 - utils 폴더 파일 (클라이언트 - 모듈)
